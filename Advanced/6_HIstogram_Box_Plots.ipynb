```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_Pol = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='Poland')].copy()

```

```python
df_DA_Pol['salary_year_avg'].plot(kind='hist', bins=25, edgecolor = 'black')

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.xlim(30000,130000)
plt.xlabel('Yearly Salary')
plt.ylabel('Count of jobs')
plt.title('Distribution od Data Analyst Salary in Poland')
plt.show()
```

```python
df_DA_Pol = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='Poland')].copy()
```

```python
df_DA_Pol['salary_year_avg'].plot(kind='box', vert = False)
```

```python
job_tit = ['Data Analyst', 'Data Scientist', 'Data Engineer']

df_Pol = df[(df['job_title_short'].isin(job_tit)) & (df['job_country']=='Poland')].copy()
df_Pol = df_Pol.dropna(subset=['salary_year_avg'])

job_list = [df_Pol[df_Pol['job_title_short'] == job_title]['salary_year_avg'] for job_title in job_tit]

plt.boxplot(job_list, labels=job_tit, vert=False)
plt.xlabel("Yearly Salary USD")
plt.title('Distribution of Salary in Poland')

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))

plt.show()
```
