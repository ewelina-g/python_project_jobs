```python
job_counts3 = df.job_title_short.value_counts().head(3)
job_counts = df.job_title_short.value_counts()
```

```python
plt.bar(job_counts3.index, job_counts3)
```

```python
plt.barh(job_counts.index, job_counts)
# jesli chcemy mieć w innej kolejności -> od największego
# musimy dodać sortowanie do serii czyli tutaj do job_counts
```

```python
plt.bar(job_counts.index, job_counts)
plt.title("Posting by Job Title")
plt.ylabel("Cunt of Job Posting")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
ef = df[['job_posted_date','salary_year_avg']].dropna(subset=['salary_year_avg']).sort_values('job_posted_date')
```

```python
ef.plot(x='job_posted_date', y='salary_year_avg', kind = 'line')
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.xlabel('Salary (USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')
plt.show()
```
